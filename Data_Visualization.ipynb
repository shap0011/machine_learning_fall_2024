### 1.3 Distribution Plot

<br>A seaborn provides a `displot()` function which is used to visualize a distribution of the univariate variable. This function uses matplotlib to plot a histogram and fit a kernel density estimate (KDE)

```python
# plot a distribution plot of 'total_bill'

# Assuming 'df' is your DataFrame containing the 'total_bill' column
sns.displot(df['total_bill'], kde=True, bins=20, color='blue')

# Add labels and title
plt.title('Distribution of Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')

# Display the plot
plt.show()
```

#### **Q6. What kind of skew does this plot have? - 1 point**

Ans: A right skew (positively skewed distribution). The mean is typically greater than the median, since the few larger values pull average to the right. The mode (the most frequent value) is usually less than both the mean and the median.

```python
# Calculate mean, median, and mode
mean_total_bill = df['total_bill'].mean()
median_total_bill = df['total_bill'].median()
mode_total_bill = df['total_bill'].mode()[0]  # mode() returns a Series, so take the first value

# Display the results
print(f"Mean: {mean_total_bill}")
print(f"Median: {median_total_bill}")
print(f"Mode: {mode_total_bill}")
```

### 1.5 Heatmap

<br> Heatmap is a two-dimensional graphical representation of data where the individual values that are contained in a matrix are represented as colors. Each square in the heatmap shows the correlation between variables on each axis.

Compute correlation between the variables using `.corr()` function.

```python
# Plot a heatmap of the correlation matrix.
# Load the dataset
df = sns.load_dataset('tips')

# Select only numeric columns
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix
correlation_matrix = numeric_df.corr()

# Create the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)

# Add labels and title
plt.title('Correlation Heatmap', fontsize=14)
plt.show()
```

```python
# count of smokers and non-smokers in the data
smoker_counts = df['smoker'].value_counts()
```

#### **Q9: Plot a bar plot bar to display count of smokers and non-smokers in the data using matplotlib - 1 point**

### 2.3 Pie Plot

Pie plot is a graphical representation of univariate data. It is a circular graph divided into slices displaying the numerical proportion. For the categorical variable, each slice of the pie plot corresponds to each of the categories.



```python
# categories in the 'day' variable
df['day'].value_counts()
```

```python
# plot the occurrence of different days in the dataset

# 'autopct' displays the percentage upto 1 decimal place
# 'radius' sets the radius of the pie plot
plt.pie(df.day.value_counts(), autopct = '%.1f%%', radius = 1.2, labels = ['Sat', 'Sun','Thur','Fri'])


# display the plot
plt.show()
```

**Observations**
* From the above pie plot, it can be seen that the data has a high proportion for Saturday followed by Sunday.

**Exploded pie plot** is a plot in which one or more sectors are separated from the disc. Use the `explode` argument to seperate the friday portion. Remeber to use `plt.show()`

#### **Q.10 Plot an exploding pie chart - 1 point**

```python
# plot the occurrence of different days in the dataset
# 'autopct' displays the percentage upto 1 decimal place
# 'radius' sets the radius of the pie plot
# display the plot

# Count the occurrences of different days
day_counts = df['day'].value_counts()

# Create the explode array to emphasize one slice (e.g., the first slice)
explode = [0.4 if i == 3 else 0 for i in range(len(day_counts))]

# Plot the pie chart
plt.figure(figsize=(8, 8))  # Set figure size
plt.pie(
    day_counts.values,
    labels=day_counts.index,
    autopct='%1.1f%%',  # Display percentages up to 1 decimal place
    explode=explode,    # Explode the first slice
    colors=['blue','orange','green','red'],  # Custom colors
    # startangle=140,     # Rotate the pie chart for better view
    radius=0.85,         # Set the radius of the pie chart
    textprops={'color': 'black', 'fontsize': 16}  # Change text color to white
)

# Add a title
plt.title('Occurrence of Different Days', fontsize=14)

# Display the plot
plt.show()
```

### 2.5 Box Plot
<br>Boxplot is a way to visualize the five-number summary of the variable. The five-number summary includes the numerical quantities like minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum. Boxplot gives information about the outliers in the data. Detecting and removing outliers is one of the most important steps in exploratory data analysis. Boxplots also tells about the distribution of the data.


#### **Q12. Plot the boxplot of `'total_bill'` to check the distribution and presence of outliers in the variable. - 1 point**

```python
# Display the plot
# plt.show()

# add labels for five number summary
# plt.text(x = 1.1, y = df['total_bill'].min(), s ='min')
# plt.text(x = 1.1, y = df.total_bill.quantile(0.25), s ='Q1')
# plt.text(x = 1.1, y = df['total_bill'].median(), s ='meadian (Q2)')
# plt.text(x = 1.1, y = df.total_bill.quantile(0.75), s ='Q3')
# plt.text(x = 1.1, y = df['total_bill'].max(), s ='max')

# add the graph title and axes labels

# display the plot

# Plot the boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, y='total_bill', color='skyblue')

# Calculate five-number summary
min_val = df['total_bill'].min()
q1 = df['total_bill'].quantile(0.25)
median = df['total_bill'].median()
q3 = df['total_bill'].quantile(0.75)
max_val = df['total_bill'].max()

# Add labels for the five-number summary
plt.text(x=0, y=min_val, s='min', color='black', fontsize=10, ha='center')
plt.text(x=0, y=q1, s='Q1', color='black', fontsize=10, ha='center')
plt.text(x=0, y=median, s='median', color='black', fontsize=10, ha='center')
plt.text(x=0, y=q3, s='Q3', color='black', fontsize=10, ha='center')
plt.text(x=0, y=max_val, s='max', color='black', fontsize=10, ha='center')

# Add a title and labels
plt.title('Boxplot of Total Bill with Five-Number Summary', fontsize=14)
plt.xlabel('1', fontsize=12)

# Display the plot
plt.show()

```

### 2.6 Subplots

`pyplot.subplot()` is a function in the matplotlib library that creates multiple subplots in a single figure. It takes three integer arguments: `numrows`, `numcols`, and `plot_number` where `plot_number` ranges from 1 to numrows * numcols

```python
# If you need multiple plots in the same canvas

# Filter the DataFrame to keep only numeric columns for the correlation matrix
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Create the subplot canvas
plt.figure(figsize=(10, 10))

# Plot 1: Heatmap of the correlation matrix
plt.subplot(2, 2, 1)
sns.heatmap(numeric_df.corr(), annot=True, cmap='YlGnBu')
plt.title('Correlation Matrix Heatmap')

# Plot 2: Countplot for the days, with hue set to 'time'
plt.subplot(2, 2, 2)
sns.countplot(x='day', data=df, hue='time')
plt.title('Count Plot of Days with Time Hue')

# Plot 3: Distribution plot of 'total_bill'
plt.subplot(2, 2, 3)
sns.histplot(df['total_bill'], kde=True)  # Updated distplot to histplot (distplot is deprecated)
plt.title('Distribution of Total Bill')

# Adjust layout
plt.tight_layout()

# Display the plots
plt.show()
```
